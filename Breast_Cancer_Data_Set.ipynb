### **1. Brief about Breast Cancer**

Breast Cancer 

It is a type of Cancer that starts in Breast. It happens because the cells insie the breast starts growing oiut of control and starts forming a lump. 

Where breast cancer starts

Breast cancers can start from different parts of the breast.

* Most breast cancers begin in the ducts that carry milk to the nipple (ductal 
cancers)
* Some start in the glands that make breast milk (lobular cancers)
* There are also other types of breast cancer that are less common like phyllodes tumor and angiosarcoma
* A small number of cancers start in other tissues in the breast. These cancers are called sarcomas and lymphomas and are not really thought of as breast cancers.

### **6.5. Explaining the Data Variables in terms of Survival and Non-Survival**

```
def box_plot(y_axis, data):

  sns.catplot(x='Status', y=y_axis, data=data, kind='box', height = 8, aspect =.7)
```

```
box_plot('Age',data)
```

**Inference** -- We can interpret that the Patients betwenn 43 - 60 survives where as those falling between 46 - 61 are tend to die.

**Inference** -- We can interpret that the no of years operations conducted between 1960 and 1966 were successfull. However that conducted between 1955 and 1965 were not that successfull. 

```
box_plot('Pos_axillary_nodes',data)
```
